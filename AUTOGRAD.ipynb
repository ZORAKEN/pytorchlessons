AUTOGRAD

AutoGrad in TensorFlow is the system that automatically computes gradients (derivatives) of operations. These gradients are essential for training neural networks because they tell the optimizer how to update model weights to reduce the loss.

```python
import tensorflow as tf

# Create a variable
x = tf.Variable(3.0)

# Record operations
with tf.GradientTape() as tape:
    y = x**2 + 2*x + 1

# Compute dy/dx
gradient = tape.gradient(y, x)

print(gradient.numpy())  # Output: 8.0
```

```python
import torch

# Inputs
x = torch.tensor(6.7)  # Input feature
y = torch.tensor(0.0)  # True label (binary)

w = torch.tensor(1.0)  # Weight
b = torch.tensor(0.0)  # Bias
# Binary Cross-Entropy Loss for scalar
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8  # To prevent log(0)
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
# Forward pass
z = w * x + b  # Weighted sum (linear part)
y_pred = torch.sigmoid(z)  # Predicted probability

# Compute binary cross-entropy loss
loss = binary_cross_entropy_loss(y_pred, y)
```

```python
loss
```

```python
# Derivatives:
# 1. dL/d(y_pred): Loss with respect to the prediction (y_pred)
dloss_dy_pred = (y_pred - y)/(y_pred*(1-y_pred))

# 2. dy_pred/dz: Prediction (y_pred) with respect to z (sigmoid derivative)
dy_pred_dz = y_pred * (1 - y_pred)

# 3. dz/dw and dz/db: z with respect to w and b
dz_dw = x  # dz/dw = x
dz_db = 1  # dz/db = 1 (bias contributes directly to z)

dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db
print(f"Manual Gradient of loss w.r.t weight (dw): {dL_dw}")
print(f"Manual Gradient of loss w.r.t bias (db): {dL_db}")
```

 1. requires_grad_(False)

This permanently tells PyTorch:

"Don't compute gradients for this tensor anymore."

```python
import torch

x = torch.tensor(2.0, requires_grad=True)

print(x.requires_grad)
```

```python
x.requires_grad_(False)

print(x.requires_grad)
```

detach()

detach() creates a new tensor that shares the same data but is not connected to the computational graph.y remembers how it was computed.
z forgets its history.
You may want the discriminator to use generated images without updating the generator:

fake_images = generator(noise)

output = discriminator(fake_images.detach())

The generator won't receive gradients from the discriminator.

```python
import torch

x = torch.tensor(2.0, requires_grad=True)

y = x * 3

z = y.detach()
```

```python
print(y.requires_grad)
print(z.requires_grad)
```

```python
print(y)
print(z)
```

3. torch.no_grad()

This temporarily disables gradient tracking inside a block of code.

```python
import torch

x = torch.tensor(2.0, requires_grad=True)

with torch.no_grad():
    y = x * 3

print(y.requires_grad)
```

```python
z = x * 4

print(z.requires_grad)
```

REDUCTION OPERATIONS

Reduction operations reduce multiple values in a tensor into one value (or one value per row/column).

```python
e = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)#creates a tensor with values low and high
e
```

```python
# sum
torch.sum(e)
# sum along columns
torch.sum(e, dim=0)
# sum along rows
torch.sum(e, dim=1)
```

```python
# mean
torch.mean(e)
# mean along col
torch.mean(e, dim=0)
# median
torch.median(e)
# max and min
torch.max(e)
torch.min(e)
```

```python
# Product of all elements.
torch.prod(e)
```

```python

# standard deviation
torch.std(e)
# variance
torch.var(e)
# argmax
torch.argmax(e)
# argmin
torch.argmin(e)
```

```python
import torch

# Model outputs (logits)
logits = torch.tensor([-2.0, -0.5, 0.0, 1.2, 3.0])

# Apply sigmoid
probabilities = torch.sigmoid(logits)

print("Probabilities:")
print(probabilities)

# Convert probabilities to True/False using a threshold of 0.5
predictions = probabilities >= 0.5

print("\nPredictions:")
print(predictions)
```

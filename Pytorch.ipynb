ARTHMETIC OPERATIONS

```python
import torch

# Scalar (0D)
a = torch.tensor(5)

# Vector (1D)
b = torch.tensor([1, 2, 3])

# Matrix (2D)
c = torch.tensor([[1, 2],
                  [3, 4]])
```

```python
x = torch.empty(2, 3)
print(x.shape)
x = torch.empty(2, 2, dtype=torch.int32)
print(x)
type(x)
```

```python
torch.empty()# fills the tensor with garbage values already in the system
torch.zeros() #erases the whiteboard and writes 0 everywhere.
torch.ones() #writes 1 everywhere.
torch.empty() #just hands you the whiteboard without erasing it.
torch.rand()#random
```

```python
import torch

x = torch.rand(2, 3)#distributed between 0 and 1.

print(x)
```

```python
torch.manual_seed(42)

print(torch.rand(3))#running again same result
```

```python
x = torch.tensor([1, 2, 3, 4])

print(x)
```

```python
# other ways

# arange ,torch.arange(start, end, step)
print("using arange ->", torch.arange(0,10,2))#tensor([0, 2, 4, 6, 8]) 10  is excluded
torch.arange(1, 6)#tensor([1, 2, 3, 4, 5])
# using linspace
print("using linspace ->", torch.linspace(0,10,10))#here no.of steps instead of stepvalue
torch.linspace(1, 5, 5)
# using eye
print("using eye ->", torch.eye(5))#creates diagonal matrixes

# using full
print("using full ->", torch.full((3, 3), 5))
```

```python
x = torch.full((3, 3), 5)#torch.full(size, fill_value)

print(x)
```
